# Análise Exploratória dos dados


```python
(titanic.groupby(['Pclass','Survived']).size() / titanic.groupby(['Pclass']).size())*100
```

```python
titanic.groupby(['Pclass','Survived','Sex']).size()
```

```python
(titanic.groupby(['Pclass','Survived','Sex']).size() / titanic.groupby(['Pclass','Survived']).size())*100
```

```python
idades = ['Recem_nascido','Crianca', 'Adolescente', 'Adulto Jovem', 'Adulto', 'Idoso']
cut_points =  [0, 5, 12, 18, 35, 60, 100]

titanic['Age_category']= pd.cut(titanic['Age'], cut_points, labels=idades)
titanic.info()
```

```python
titanic.groupby(by = 'Age_category').size()
```

```python
(titanic.groupby(['Age_category','Survived']).size() / titanic.groupby('Age_category').size())*100
```

```python
(titanic.groupby(['Age_category','Survived','Sex']).size() / titanic.groupby(['Age_category','Survived']).size())*100
```

```python
age_survived = sns.FacetGrid(titanic, col='Survived')
age_survived.map(sns.distplot, 'Age')
```

```python
corrMatrix = titanic[['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'SibSp', 'Parch']].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```
